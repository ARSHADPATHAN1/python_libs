```python
x = [1,5,4]
y = [23,45,89]
x2 = [3,6,9]
y2 = [4,35,79]
```

```python
plt.plot(x,y)
plt.plot(x2,y2)
```

# to change the style of graph/plot


```python
from matplotlib import style
```

```python
style.use('ggplot')
plt.plot(x,y)
plt.plot(x2,y2)
```

# can see more styles in https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html

```python
style.use('ggplot')
plt.plot(x,y ,label="first")
plt.plot(x2,y2 ,label="second")
plt.legend()
```

```python
style.use('ggplot')
plt.plot(x,y ,label="first" , linewidth=4)
plt.plot(x2,y2 ,label="second")
plt.legend()
```

```python
style.use('ggplot')
plt.hist(x,y ,label="first")
plt.plot(x2,y2 ,label="second")
plt.legend()
```

```python
style.use('ggplot')
plt.hist(x ,label="first")
plt.plot(x2,y2 ,label="second")
plt.legend()
```
